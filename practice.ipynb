```python
dt=pd.read_csv("dataTree1.csv")
d={"UK":0,"USA":1,"N":2}
d1={"YES":1,"NO":0}
dt["Nationality"]=dt["Nationality"].map(d)
dt["Go"]=dt['Go'].map(d1)
dt.head()
dt.iloc[:11,4]
```

```python
dtree=DecisionTreeClassifier(criterion="gini")
dtree.fit(dt.iloc[:11,:4],dt.iloc[:11,4])
tree.plot_tree(dtree,feature_names=dtree.feature_names_in_,filled=True,rounded=True,proportion=True)
```
